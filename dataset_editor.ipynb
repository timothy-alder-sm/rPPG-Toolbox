```python
import glob
import os
import zipfile
import fnmatch

def unzip_all(data_path, sub_dir="*", unzip_RGB=True, unzip_NIR=False, delete_zip=False):
    directories = glob.glob(data_path + os.sep + "subject*" + os.sep + sub_dir)
    for dir_path in directories:   
        for root, dirs, files in os.walk(dir_path):
            for filename in fnmatch.filter(files, '*.zip'):
                if unzip_NIR == False and filename == "NIR.zip":
                    continue
                if unzip_RGB == False and filename == "RGB.zip":
                    continue
                
                zip_path = os.path.join(root, filename)
                print(zip_path)
                
                zipfile.ZipFile(zip_path).extractall(root)
                
                if delete_zip:
                    os.remove(zip_path)

unzip_all("D:\\MR-NIRP", "*_driving_small_motion_975")
```

```python
def delete_zip(data_path, sub_dir="*", delete_RGB=False, delete_NIR=False):
    directories = glob.glob(data_path + os.sep + "subject*" + os.sep + sub_dir)
    for dir_path in directories:   
        for root, dirs, files in os.walk(dir_path):
            for filename in fnmatch.filter(files, '*.zip'):
                if filename != "NIR.zip" and filename != "RGB.zip":
                    continue
                if delete_NIR == False and filename == "NIR.zip":
                    continue
                if delete_RGB == False and filename == "RGB.zip":
                    continue
                
                zip_path = os.path.join(root, filename)
                print(zip_path)
                
                os.remove(zip_path)
                
delete_zip("D:\\MR-NIRP", delete_NIR=True)
```
